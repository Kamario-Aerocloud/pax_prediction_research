```python
# # Connect to MongoDB via forwarded port
# client = MongoClient(
#     "mongodb://localhost:8888/"
#     "?directConnection=true"
# )

# # Access your database and collection
# db = client["aeroclouddb"]
# collection = db["flights"]

# Fetch some data
#data = list(collection.find())

#df = pd.DataFrame(data)

df1 =  pd.read_csv('Datasets/augmented_SRQ_data_v3.csv')
df2 =  pd.read_csv('Datasets/aeroclouddb.flights.csv')

df1['flight_number_clean'] = df1['Flight'].str[2:]
df1['Date'] = pd.to_datetime(df1['Date'])

df2['Date'] = pd.to_datetime(df2['scheduled_date'])
df2['Date'] = df2['Date'].dt.strftime('%Y-%m-%d')

flights_df = df1.copy()
extra_info_df= df2.copy()

flights_df
```

```python
extra_info_df
```

# Merge datasets

```python
# Clean and convert flight number in df1
flights_df['flight_number_clean'] = flights_df['flight_number_clean'].astype(str)

# Convert df2 flight_number to string
extra_info_df['flight_number'] = extra_info_df['flight_number'].astype(str)
extra_info_df['Date'] = pd.to_datetime(extra_info_df['Date'])

# Merge on cleaned flight number and scheduled_date
merged_df = pd.merge(
    flights_df,
    extra_info_df,
    left_on=['Date', 'flight_number_clean'],
    right_on=['Date', 'flight_number'],
    how='inner'
)

merged_df
```

```python
merged_df.info()
```

```python
merged_df.to_csv('SRQ_flights.csv', index=False)
```
